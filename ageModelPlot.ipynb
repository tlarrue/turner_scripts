```python
%matplotlib inline
import numpy as np
import os, sys
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
topDir = '/projectnb/trenders/proj/turner/outputs/age_historyvar_regression'
```

```python
#RMA MODEL RESULTS - using 5x5 median filtered maps & 20,000 pts

rmaCSV = os.path.join(topDir, 'regression_models/RMA_results_halfhalf_5x5med_20000pt_11172014.csv')
inputFile = open(rmaCSV, 'rb')
inputData = np.genfromtxt(inputFile, delimiter=',', names=True, case_sensitive=False, dtype=None) #structured array of strings
inputFile.close()

y_fields = ['PREDICTION','RESIDUALS', 'CCA_INDEX']
x_field = 'TARGET'

x = inputData[x_field]
plotNum = 0
plt.figure(figsize=(16,4))
for i in y_fields:
    plotNum += 1
    plt.subplot(1,3,plotNum)
    y = inputData[i]
    plt.hist2d(x, y, bins=50, norm=LogNorm())
    plt.colorbar()
    plt.title(i + " vs OBSERVED AGE")
print 'MEAN SQUARE ERROR:', inputData['MEAN_SQ_ERR'][0]
print 'RMSE:', float(inputData['MEAN_SQ_ERR'][0])**(1./2.)
```
